### 3. Shear

```python
# -- c. Get the image dimensions ────────────────────
w, h = img.size
print("Width:", w)
print("Height:", h)
```

```python
# -- d. define the shear matrix ──────────────────────────
# Choose X-axis or Y-axis shear. The shear factor controls how much the image slants — start with 0.5 then experiment.
shear_factor = 0.5

# X-axis shear
matrix = (1, shear_factor, 0,
          0, 1, 0)
```

```python
# -- e. Apply the shear transformation to the image ──────────────────────────
# PIL's transform() takes the inverse affine matrix:
# [ 1    shx   tx ]
# [ shy  1     ty ]
new_width = int(w + abs(shear_factor) * h)

img_sheared = img.transform(
    (new_width, h),
    Image.AFFINE,
    matrix,
    resample=Image.BICUBIC
)
```

```python
# -- f. Save the sheared image(The new image name should be "task1_3_sheared.jpg") 

img_sheared.save("task1_3_sheared.jpg")
img_sheared.show()
```

```python
# Experiment 1
#Change shear factor from 0.5 to 0.1, 0.3, 0.8 and compare results

factors = [0.1, 0.3, 0.8]

for f in factors:
    new_width = int(w + abs(f) * h)

    matrix = (1, f, 0,
              0, 1, 0)

    img_shear = img.transform(
        (new_width, h),
        Image.AFFINE,
        matrix,
        resample=Image.BICUBIC
    )

    img_shear.save(f"shear_{f}.jpg")
    img_shear.show()
```

```python
# Experiment 2:
#Switch from X-axis to Y-axis shear matrix

shear_factor = 0.5

new_height = int(h + abs(shear_factor) * w)

matrix = (1, 0, 0,
          shear_factor, 1, 0)

img_shear_y = img.transform(
    (w, new_height),
    Image.AFFINE,
    matrix,
    resample=Image.BICUBIC
)

img_shear_y.save("shear_Y_axis.jpg")
img_shear_y.show()
```

```python
# Experiment 3
#Update the canvas multiplier to match your new factor

shear_factor = 0.8

new_width = int(w + abs(shear_factor) * h)

matrix = (1, shear_factor, 0,
          0, 1, 0)

img_shear_canvas = img.transform(
    (new_width, h),
    Image.AFFINE,
    matrix,
    resample=Image.BICUBIC
)

img_shear_canvas.save("shear_canvas_adjusted.jpg")
img_shear_canvas.show()
```

```python
# Experiment 4
#Try combining X and Y shear in one matrix

shx = 0.5
shy = 0.3

new_width = int(w + abs(shx) * h)
new_height = int(h + abs(shy) * w)

matrix = (1, shx, 0,
          shy, 1, 0)

img_shear_both = img.transform(
    (new_width, new_height),
    Image.AFFINE,
    matrix,
    resample=Image.BICUBIC
)

img_shear_both.save("shear_xy.jpg")
img_shear_both.show()
```

# Intensity Transformations
Negative · Log · Power Law (Gamma)

```python

# ── 1. Negative ──────────────────────────────────── 
# Method 1: NumPy array manipulation

img_np = np.array(img)

negative_np = 255 - img_np

negative_img_np = Image.fromarray(negative_np.astype(np.uint8))
negative_img_np.show()

```

```python
# Method 2: PIL's ImageOps
from PIL import Image, ImageOps
negative_img_ops = ImageOps.invert(img)
negative_img_ops.show()
negative_img_np.save("task2_1_negative_numpy.jpg")
negative_img_ops.save("task2_1_negative_imageops.jpg")
```

```python
# ── 2. Log transformation ──────────────────────────

# s = c · log(1 + r) --> We need to find c.
# We want the maximum output value (s) to be 255 when the maximum input value (r) is 255:
# 255 = c · log(1 + 255)
# c = 255 / log(1 + 255)


c = 255 / np.log(1 + 255)
print("c =", c)

img_np = np.array(img, dtype=np.float32)

log_transformed = c * np.log(1 + img_np)

log_transformed = np.clip(log_transformed, 0, 255)
log_transformed = log_transformed.astype(np.uint8)

log_img = Image.fromarray(log_transformed)
log_img.show()

log_img.save("task2_2_log.jpg")

# Apply the log transformation to each pixel
```

```python

# ── 3. Power-law / Gamma correction ───────────────
gamma = 0.5   # قيمة أقل من 1 تزيد الإضاءة/التباين
c = 1

img_np = np.array(img, dtype=np.float32) / 255.0
```

```python
# Apply gamma transformation to each pixel
power_transformed = c * np.power(img_np, gamma)

power_transformed = np.clip(power_transformed * 255, 0, 255)
power_transformed = power_transformed.astype(np.uint8)

power_img = Image.fromarray(power_transformed)
power_img.show()

power_img.save("task2_3_power.jpg")
```
